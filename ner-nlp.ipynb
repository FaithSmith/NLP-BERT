```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from simpletransformers.ner import NERModel, NERArgs
data = pd.read_csv("../input/entity-annotated-corpus/ner_dataset.csv", encoding='latin-1', usecols=['Sentence #','Word','Tag'])
data.head(20)
```

```python
data["Sentence #"].fillna(method='ffill', inplace=True)
data["Sentence #"] = LabelEncoder().fit_transform(data["Sentence #"] )
data.head(10)
```

```python
data.rename(columns={"Sentence #":"sentence_id","Word":"words","Tag":"labels"}, inplace =True)
data["labels"] = data["labels"].str.upper()
X= data[["sentence_id","words"]]
Y =data["labels"]
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size =0.2, random_state=42)
#building up train data and test data
train_data = pd.DataFrame({"sentence_id":x_train["sentence_id"],"words":x_train["words"],"labels":y_train}).reset_index()
test_data = pd.DataFrame({"sentence_id":x_test["sentence_id"],"words":x_test["words"],"labels":y_test}).reset_index()
train_data
```

```python
labels = data["labels"].unique().tolist()
labels
```

```python
model_type = 'bert'
model_name = 'bert-base-cased'
```

```python
args = NERArgs()
args.num_train_epochs = 6
args.learning_rate = 1e-4
args.overwrite_output_dir =True
args.train_batch_size = 32
args.eval_batch_size = 32
args.output_dir = f'/kaggle/working/outputs/{model_type}/'
args.best_model_dir = f'/kaggle/working/outputs/{model_type}/best_model'
# wandb_kwargs = {"name": model_name}
# wandb_project = "Named Entity Recognition"
args.evaluate_during_training = True
```

```python

model = NERModel(model_type, 'bert-base-cased',labels=labels,args =args)
```

```python
model.train_model(train_data,eval_data = test_data)
```

```python
result, model_outputs, preds_list = model.eval_model(test_data)
```

```python
result
```

```python
prediction, model_output = model.predict(["What is the new name of Bangalore"])
```
